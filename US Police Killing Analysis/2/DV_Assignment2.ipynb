## 2: Plot 12 most dangerous cities

```python
city_freq = df.city.value_counts().head(12)
city_freq
```

```python
fig, ax =plt.subplots(figsize=(8,5))
x = city_freq.index
y = city_freq.values
ax = sns.barplot(x, y)
ax.set_title("Most dangerous cities",color="b",size=18)
ax.set_xticklabels(city_freq.index,rotation=45);
```

```python
fig, ax =plt.subplots(figsize=(8,5))
ax= sns.countplot(x=df.city, order=city_freq.index,data=df)
ax.set_title("Most dangerous cities",color="b",size=18)
ax.set_xticklabels(city_freq.index,rotation=45);
```

## 3: Plot 20 most dangerous states

```python
state_freq = df.state.value_counts().head(20)
state_freq
```

```python
fig, ax =plt.subplots(figsize=(8,5))
x = state_freq.index
y = state_freq.values
ax = sns.barplot(x, y)
ax.set_title("Most dangerous states",color="b",size=18)
ax.set_xticklabels(x,rotation=45);
```

```python
fig, ax =plt.subplots(figsize=(8,5))
ax= sns.countplot(x=df.state, order=state_freq.index,data=df)
ax.set_title("Most dangerous states",color="b",size=18)
ax.set_xticklabels(state_freq.index,rotation=45);
```

## 4: Plot Having mental ilness or not for killed people

```python
illness = df.signs_of_mental_illness.value_counts()
illness
```

```python
fig, ax = plt.subplots()
ax=sns.countplot(x="signs_of_mental_illness",data=df)
for p in ax.patches:
    ax.annotate(p.get_height(),(p.get_x() + 0.32, p.get_height() + 10))
```

## 5: Plot number of Threat Types
